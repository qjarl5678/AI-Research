```python
from cProfile import label

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

column_names = ['idx', '연차', '연봉']
df = pd.read_csv('./dataset/Salary_dataset.csv', names=column_names, skiprows=1)

print(df.head())
```

```python
# 독립 변수와 종속 변수 분리
x = df['연차'].values.reshape(-1, 1) # (n, 1) 형태로 변환
y = df['연봉']

# 선형 회귀 모델 초기화 및 학습
model = LinearRegression()
model.fit(x, y)

# predict
y_pred = model.predict(x)


# 맑은 고딕 폰트를 matplotlib 기본 폰트로 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지

# visualization
plt.scatter(df['연차'], df['연봉'], color='blue', label = '실제 데이터')
plt.plot(df['연차'], y_pred, color='red', label = '회귀 라인')
plt.xlabel("연차")
plt.ylabel("연봉")
plt.title("선형회귀: 연차별 연봉")
plt.legend()
plt.show()
```

```python
# y = mx + b
# y: 종속 변수 (예측값, 예: 연봉)
# x: 독립 변수 (입력값, 예: 경력 연수)
# m: 기울기 (Independent Variable이 변할 때 y가 얼마나 변하는지 나타냄)
# b: 절편 (직선이 y-축과 만나는 점)

# 모델 평가
print("Slope (m):", model.coef_[0])
print("Intercept (b):", model.intercept_)
print("Mean Squared Error (MSE):", mean_squared_error(y, y_pred))
print("R-squared:", r2_score(y, y_pred))
```
